```python
#importação das bibliotecas necessárias para essa análise
import pysus.online_data.CNES as CNES
import pandas as pd
import pandasql as ps
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Criando funcção para automatizar a coleta de dados já que só é possível baixar mês/ano de cada vez
def downloadDf(year,month,params={
    'group':'LT',
    'state':'GO'
}):
    df = CNES.download(params['group'],params['state'],year,month)
    df = pd.DataFrame(df) 
    return df
```

***Para esta análise devemos considerar o inicio e o fim da pandemia no Brasil, vamos considerar as seguintes datas:***
- Início da pandemia: 20 de março de 2020 - Decreto de estado de calamidade (http://www.planalto.gov.br/ccivil_03/portaria/DLG6-2020.htm).
- Fim da pandemia: 20 de maio de 2022 - Decreto de revogação do decreto de estado de calamidade (http://www.planalto.gov.br/ccivil_03/_ato2019-2022/2022/Decreto/D11077.htm#art1)

## Análise da evolução do número de leitos no tempo

***Vamos separar os dados em 3 períodos: antes durante e depois da pandemia utilizando as datas selecionadas anteriormente: 20 de março de 2020 a  20 de maio de 2022.***

```python
# alterando os tipo de dados para inteiro para realizar as comparações
dataset = dataset.astype({
    'COMPETEN': 'int32',
    'ANO':'int32',
    'MES':'int32'
})
```

```python
# Criando coluna categórica referente aos periodos da pandemia
query = """
        SELECT 
               CODUFMUN,
               QT_EXIST,
               QT_CONTR,
               QT_SUS,
               QT_NSUS,
               COMPETEN,
               ANO,
               MES,
               CASE
                   WHEN COMPETEN BETWEEN 202004  AND 202205 THEN '2 - DURANTE'
                   WHEN COMPETEN <= 202003 THEN '1 - ANTES'
                   ELSE '3 - DEPOIS'
                END PERIODO
               
        FROM dataset ds

        """

dataset_periodos = pd.DataFrame(ps.sqldf(query,locals())) 
dataset_periodos
```

```python
# Criando sumarizando a quantidade de leitos por período
query = """
        SELECT 
               PERIODO,
               SUM(QT_EXIST) SUM_QT_EXIST
               
        FROM dataset_periodos dsp
        group by PERIODO

        """

agrupado_periodo = pd.DataFrame(ps.sqldf(query,locals())) 
agrupado_periodo
```

```python
#visualização gráfica da evolução da quantidade de leitos no tempo
plt.figure(figsize=(15,8))
sns.barplot(data=agrupado_periodo, x="PERIODO", y="SUM_QT_EXIST").set(xlabel='Períodos de Pandemia', ylabel='Quantidade de Leitos' )
```

***Aqui observamos gráficamente que durante a pandemia houve um crescimento de número de leitos em relação ao período antes da pandemia e também vemos o declínio de número de leitos no período depois da pandemia em relação ao período durante a pandemia, mas o número de leitos no período depois da pandemia, apesar do grande delclíneo, ainda é maior que o número de leitos no período anes da pandemia.***

## Distribuição geográfica dos leitos por município

```python
#sumarizando os municipios e contando a quantidade de ocorrencias no nosso dataset
agrupado_codmun = dataset_periodos.groupby(['CODUFMUN']).size()
agrupado_codmun
```

***De acordo com a informação acima temos 194 municípios diferentes em nosso dataset, por isso vou fazer um subconjunto com os 10 municípios que possuem mais leitos e visualizar a evolução de leitos por período.***

```python
# Dataframe com os 10 municípios com maior número de leitos
query = """
        SELECT
        *
        FROM
        dataset_periodos
        WHERE CODUFMUN IN  (SELECT CODUFMUN 
                            FROM ( SELECT 
                                         CODUFMUN,
                                         sum(QT_EXIST) SUM_QT_EXIST
                                   FROM dataset_periodos dsp
                                   group by CODUFMUN
                                   ORDER BY 2 DESC
                                   LIMIT 10
                                   ) --TOP 10
                            )
    

        """

dataset_top10 = pd.DataFrame(ps.sqldf(query,locals())) 
dataset_top10
```

***Acima temos um dataframe somente com dados referente aos 10 municípios com maior número de leitos. Observamos que o dataset completo tem 106443 linhas mas esse subconjunt tem 53323.***

```python
# Sumarizando os 10 municípios com maior número de leitos
query = """
        SELECT
        dsp.CODUFMUN,
        sum(dsp.QT_EXIST) SUM_QT_EXIST,
        sum(dsp.QT_SUS) SUM_QT_SUS, 
        sum(dsp.QT_NSUS) SUM_QT_NSUS,
        dsp.PERIODO
        FROM
        dataset_periodos dsp
        WHERE dsp.CODUFMUN IN  (SELECT CODUFMUN 
                            FROM ( SELECT 
                                         CODUFMUN,
                                         sum(QT_EXIST) SUM_QT_EXIST
                                   FROM dataset_periodos dsp
                                   group by CODUFMUN
                                   ORDER BY 2 DESC
                                   LIMIT 10
                                   ) --TOP 10
                            )
        GROUP BY dsp.CODUFMUN, dsp.PERIODO
    

        """

agrupado_top10 = pd.DataFrame(ps.sqldf(query,locals())) 
agrupado_top10
```

```python
#vusualização gráfica da evolução da quantidade de leitos dos 10 municípios com maior número de leitos por período da pandemia
plt.figure(figsize=(15,8))
sns.barplot(data=agrupado_top10, x="CODUFMUN", y="SUM_QT_EXIST",hue="PERIODO",palette='viridis',errorbar=None,).set(xlabel='Código de municípios', ylabel='Quantidade de Leitos' )
```

***No gráfico acima verificamos que entre os 10 municípios com maior número de leitos, que todos seguiram a tendência observada no gráfico anterior e mostram aumento na quantidade de leitos.***

## Quantidade de leitos SUS e leitos não SUS no tempo e espaço

```python
#criando visualização gráfica comparativa dos leitos sus e leitos não sus
#utilizamos o dataframe dos 10 municípios com maior número de leitos para melhorar a visualização
#gráfico mostra separação por períodos da pandemia
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(15,8))
fig.suptitle('Análise dos Leitos por Municípios')
axes[0].set(title='Leitos SUS' )
axes[1].set(title='Leitos Não SUS')
sns.barplot(ax=axes[0],data=agrupado_top10,x="CODUFMUN",y="SUM_QT_SUS",hue="PERIODO",palette='magma',errorbar=None)
sns.barplot(ax=axes[1],data=agrupado_top10,x="CODUFMUN",y="SUM_QT_NSUS",hue="PERIODO",palette='viridis',errorbar=None)
axes[0].set(xlabel='Código de municípios', ylabel='Quantidade de Leitos')
axes[1].set(xlabel='Código de municípios', ylabel='Quantidade de Leitos')
```
